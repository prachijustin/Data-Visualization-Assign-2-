<h3>Converting report_date to Timestamp format</h3>

```python
data['report_date'] = pd.to_datetime(data.report_date, format='%Y-%m-%d', errors='coerce')
```

```python
type(data.report_date[0])
```

<h3>Removing NULL columns</h3>

```python
data.isnull().sum()
```

```python
data = data.dropna(how='all',axis=1)
data.head()
```

<h3>Filling NaN(if any) with 0</h3>

```python
data.fillna(0)
```

<h3>Q1. Show data field code 'AR0005' values for 'Argentina-Catamarca' for entire duration </h3>

```python
Catamarca = data[(data.location == 'Argentina-Catamarca') & (data.data_field_code == 'AR0005')]
Catamarca.head()
```

```python
Catamarca.value = Catamarca.value.astype(int)
```

```python
Catamarca_days = Catamarca.report_date
Catamarca_data = Catamarca.value.groupby(Catamarca_days).sum()
```

```python
plt.figure(figsize=(12,7))
plt.plot(Catamarca_days, Catamarca_data)
plt.xlabel('Days')
plt.ylabel('Values')
plt.title('Catamarca Values for entire duration')
```

<h3>Q2. Plot 'Argentina-CABA' and 'Argentina-Catamarca' for number of cases values for entire duration to identify the difference.</h3>

```python
CABA = data[(data.location == 'Argentina-CABA') & (data.unit == 'cases')]
```

```python
CABA.value = CABA.value.astype(int)
```

```python
caba_dates = CABA.report_date
cabas = CABA.value.groupby(caba_dates).sum()
```

```python
Catama = data[(data.location == 'Argentina-Catamarca') & (data.unit == 'cases')]
Catama.value = Catama.value.astype(int)
```

```python
catamar_dates = Catama.report_date
catamar = Catama.value.groupby(catamar_dates).sum()
```

```python
plt.figure(figsize=(12,7))
plt.plot(cabas.index, cabas, 'b', label='Argentina-CABA')
plt.plot(catamar.index, catamar, 'r', label='Argentina-Catamarca')
plt.xlabel('Dates')
plt.ylabel('Values')
plt.title('Number of case values for whole duration')
plt.legend()
plt.show()
```
